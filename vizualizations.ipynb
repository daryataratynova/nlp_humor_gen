```python
kaggle = pd.read_csv('kaggle_dataset.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
distribution = kaggle['news_tag'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(distribution.index, distribution.values, color='skyblue')
plt.xlabel('News Tag')
plt.ylabel('Distribution')
plt.title('Distribution of News Tags')
plt.xticks(rotation=45)
plt.show()
```
